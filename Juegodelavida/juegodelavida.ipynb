```python
class adt2d():
    def __init__(self, renglon, columna):
        self.__columnas__ = columna
        self.__renglones__ = renglon
        self.__data__ = [[0 for i in range(columna)] for i in range (renglon)]
    def to_string(self):
        for i in range(self.__renglones__):
            print(self.__data__[i])
    def clear(self, dato):
        for i in range(self.__renglones__):
            for x in range(self.__columnas__):
                self.__data__[i][x] = dato
    def get_row_size(self):
        return(self.__renglones__)
    def get_col_size(self):
        return (self.__columnas__)

    def set_item(self,renglon, columna, dato):
        self.__data__[renglon][columna] = dato

    def get_item(self,renglon,columna):
        return self.__data__[renglon][columna]
```

```python

class Juegodelavida():
    def __init__(self, renglones, columnas, lista_de_poblacion_inicial, generaciones):
        self.__copytablero__ = adt2d(renglones, columnas)
        self.__tablero__ = adt2d(renglones, columnas)
        self.__columnas__ = columnas
        self.__renglones__ = renglones
        self.__generaciones__ = generaciones
        self.__gen__ = 0
        self.__copytablero__.clear("M")
        self.__tablero__.clear("M")
        for par in lista_de_poblacion_inicial:
            self.__tablero__.set_item(par[0], par[1], 'V')
    
    def to_string(self):
        print(f"----- GENERACION {self.__gen__} ------")    
        return self.__tablero__.to_string()

    def get_vecinos(self, renglon, columna):
        vecinos = []
        for ren in range(renglon-1, renglon+2, 1):
            for col in range(columna-1 , columna+2,1):
                if ren < 0 or ren >= self.__tablero__.get_row_size() or col < 0 or col >= self.__tablero__.get_col_size() or (ren==renglon and col==columna):    
                    pass
                else:
                        vecinos.append((ren,col))
        return vecinos

    def get_vecinos_vivos(self, renglon, columna):
        vivos = 0
        vecinos = self.get_vecinos(renglon, columna)
        for par in vecinos:
            if self.__tablero__.get_item(par[0],par[1]) == 'V':
                vivos += 1
        return vivos

    def aplicar_reglas(self):
        self.to_string()
        for generaciones in range(self.__generaciones__):
            for ren in range(self.__renglones__):
                for col in range(self.__columnas__):
                    if self.__tablero__.get_item(ren, col) == 'V':           
                        if self.get_vecinos_vivos(ren,col) in [2,3]:
                            pass
                        elif self.get_vecinos_vivos(ren,col) in [0,1]:
                            self.__tablero__.set_item(ren, col, 'M')
                        elif self.get_vecinos_vivos(ren,col) > 4:
                            self.__tablero__.set_item(ren, col, 'M')
                    else:
                        if self.get_vecinos_vivos(ren,col) == 3:
                                self.__tablero__.set_item(ren, col, 'V')
            self.__gen__ = generaciones + 1
            self.to_string()


```

```python
juego = Juegodelavida(6,6,[(1,1),(1,2),(1,3),(2,3)],5)
print("  ")
juego.aplicar_reglas()

```
